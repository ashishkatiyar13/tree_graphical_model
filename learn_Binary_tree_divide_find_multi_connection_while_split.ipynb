```python
def check_leaf_prec(prec, node):
    n = prec.shape[0]
    count = 0
    for i in range(n):
        if prec[i, node] !=0:
            count = count +1
        if count > 2:
            return False
    return True
```

```python
def exchange_leaf_internal(prec, node):
    n = prec.shape[0]
    for i in range(n):
        if prec[i, node] !=0:
            if check_leaf_prec(prec, i):
                break
    res_row = prec[i, :]
    res_col = prec[:, i]
    prec[i, :] = prec[node, :]
    prec[:, i] = prec[:, node]
    prec[node, :] = res_row
    prec[:, node] = res_col
    return prec
            
```
